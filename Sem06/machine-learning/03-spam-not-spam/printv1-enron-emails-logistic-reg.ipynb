```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv("./spam_ham_enron.csv")
df
```

```python
df.tail()
```

```python
data = df[["text", "label_num"]]
data
```

```python
data = data.rename(columns={"text": "message", "label_num": "label"})
data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data['message'], data['label'], test_size=0.2, random_state=1)
```

```python
print(data.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
count_vector = CountVectorizer()
training_data = count_vector.fit_transform(X_train).toarray()
```

```python
training_data
```

```python
testing_data = count_vector.transform(X_test).toarray()
testing_data
```

```python
frequency_matrix = pd.DataFrame(training_data, columns=count_vector.get_feature_names_out())
frequency_matrix.head()
```

#### **Training Logistic Regression Model**

```python
clf = LogisticRegression(random_state=0).fit(training_data, Y_train)
```

```python
predictions = clf.predict(testing_data)
predictions
```

#### **Evaluating the Model**

```python
print("Accuracy score: ", format(accuracy_score(Y_test, predictions)))
print("Precision score: ", format(precision_score(Y_test, predictions)))
print("Recall score: ", format(recall_score(Y_test, predictions)))
print("F1 Score: ", format(f1_score(Y_test, predictions)))
print("Confusion Matrix: ", format(accuracy_score(Y_test, predictions)))
```
